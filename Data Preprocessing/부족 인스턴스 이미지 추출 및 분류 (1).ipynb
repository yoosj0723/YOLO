```python
import os
import shutil

# 원본 이미지와 라벨이 저장된 디렉토리
image_dir = 'C:/Users/rund6/Desktop/YOLOv8n_ysj/최종 데이터_copy/dataset/export/image'  # 이미지 디렉토리 경로
label_dir = 'C:/Users/rund6/Desktop/YOLOv8n_ysj/최종 데이터_copy/dataset/export/labels'  # 라벨 디렉토리 경로

# 클래스 인덱스와 디렉토리 경로를 매핑
class_dirs = {
    5: 'C:/Users/rund6/Desktop/부족인스턴스분류/refrigerator_xl2',
    10: 'C:/Users/rund6/Desktop/부족인스턴스분류/allinone_air_conditioner',
    11: 'C:/Users/rund6/Desktop/부족인스턴스분류/solar_panel',
    18: 'C:/Users/rund6/Desktop/부족인스턴스분류/vending_machine_s',
    19: 'C:/Users/rund6/Desktop/부족인스턴스분류/vending_machine_l',
    24: 'C:/Users/rund6/Desktop/부족인스턴스분류/dehumidifier'
}

# 라벨 파일을 읽어 각 클래스에 해당하는 이미지와 라벨을 복사
for label_file in os.listdir(label_dir):
    if label_file.endswith('.txt'):
        label_path = os.path.join(label_dir, label_file)
        with open(label_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                class_index = int(line.split()[0])
                if class_index in class_dirs:
                    image_file = label_file.replace('.txt', '.png')
                    image_path = os.path.join(image_dir, image_file)
                    if os.path.exists(image_path):
                        # 해당 클래스의 디렉토리로 파일 복사
                        print(f'Copying {image_file} and {label_file} to {class_dirs[class_index]}')
                        shutil.copy(image_path, class_dirs[class_index])
                        shutil.copy(label_path, class_dirs[class_index])
                    else:
                        print(f"Image not found: {image_path}")
                    break  # 해당 파일에 대해 이미 복사했으므로 루프를 나간다.

print("분류가 완료되었습니다.")
```

```python
import os
from collections import defaultdict

# 라벨이 저장된 디렉토리
label_dir = 'C:/Users/rund6/Desktop/YOLOv8n_ysj/최종 데이터_copy/dataset/export/labels'

# 클래스별 개수를 저장할 딕셔너리
label_counts = defaultdict(int)

# 라벨 파일을 읽어 각 클래스의 개수를 셈
for label_file in os.listdir(label_dir):
    if label_file.endswith('.txt'):
        label_path = os.path.join(label_dir, label_file)
        with open(label_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                class_index = int(line.split()[0])
                label_counts[class_index] += 1

# 각 클래스의 개수 출력
for class_index, count in label_counts.items():
    print(f"Class {class_index}: {count} instances")

print("모든 라벨의 개수를 세었습니다.")
```
